```python
# Load numerical and plotting libraries
import numpy as np
import matplotlib.pyplot as plt
# Enable inline plotting in Jupyter notebooks
%matplotlib inline
```

```python
# Generate uniform random numbers in [0, 1)
N = 10000
seed = 42               # fixed seed for reproducibility
rng = np.random.default_rng(seed)  # create generator
data = rng.random(N)    # generate N random numbers
 
# Plot histogram: 20 bins in [0, 1)
plt.hist(data, bins=20, range=(0, 1));
```

```python
def below_fraction(num, seed = 42):
  """Compute the fraction of numbers below 0.5
     num  : int, number of random samples
     seed : int, random seed for reproducibility, default 42"""
  rng = np.random.default_rng(seed)
  data = rng.random(num)
  below = np.count_nonzero(data < 0.5)
  return below / num

Ns = [1, 10, 100, 1000, 10000, 100000]         # different sample sizes
ratios = [below_fraction(n) for n in Ns]       # compute ratios
plt.plot(Ns, ratios, marker='o')        # plot results
plt.axhline(0.5, linestyle='--', c='r') # expected value
plt.xscale('log')                       # log scale for x-axis 
```

```python
# Estimate pi
N = [1, 10, 100, 1000, 10000, 100000, 1000000]
seed_x = 15
seed_y = 25

rng_x = np.random.default_rng(seed_x)
rng_y = np.random.default_rng(seed_y)

delta = []

for i in range(7):
    data_x = rng_x.random(N[i])
    data_y = rng_y.random(N[i])
    r = np.sqrt((data_x - 0.5)**2 + (data_y - 0.5)**2)
    below = np.count_nonzero(r < 0.5)
    pi = below / N[i] * 4
    i += 1
    delta.append(pi - np.pi)

plt.plot(N, delta, marker='o')       # plot results
plt.axhline(0, linestyle='--', c='r') # expected value
plt.xscale('log')

```
